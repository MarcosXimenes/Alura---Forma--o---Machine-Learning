```python
import pandas as pd

df_exames = pd.read_csv(
    "https://raw.githubusercontent.com/alura-cursos/reducao-dimensionalidade/master/data-set/exames.csv"
)
df_exames
```

```python
from sklearn.model_selection import train_test_split
from numpy import random

SEED = 123143
random.seed(SEED)

exames = df_exames.drop(columns=["id", "diagnostico"])
exames_1 = exames.drop(columns=["exame_33"])
diagnostico = df_exames.diagnostico

X_train, X_test, y_train, y_test = train_test_split(exames_1, diagnostico)
display(X_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc_1 = RandomForestClassifier(n_estimators=100, random_state=SEED)
rfc_1.fit(X_train, y_train)
print("Resultado da classificação %.2f%%" % (rfc_1.score(X_test, y_test) * 100))
```

```python
from sklearn.dummy import DummyClassifier

dummy = DummyClassifier(strategy="most_frequent")
dummy.fit(X_train, y_train)
print(
    "Resultado da Classificação (Dummy): %.2f%%" % (dummy.score(X_test, y_test) * 100)
)
```

```python
df2plot = pd.melt(
    pd.concat([diagnostico, exames_1], axis=1),
    id_vars="diagnostico",
    var_name="exames",
    value_name="valores",
)
df2plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 5))
plt.xticks(rotation=90)
sns.violinplot(x="exames", y="valores", hue="diagnostico", data=df2plot, split=True)
```

```python
from sklearn.ensemble import RandomForestClassifier

exames_3 = exames_2.drop(columns=["exame_29", "exame_4"])
X_train, X_test, y_train, y_test = train_test_split(exames_3, diagnostico)

rfc_2 = RandomForestClassifier(n_estimators=100, random_state=SEED)
rfc_2.fit(X_train, y_train)
print("Resultado da classificação %.2f%%" % (rfc_2.score(X_test, y_test) * 100))
```

```python
def classificar(exames):
    SEED = 123143
    random.seed(SEED)
    X_train, X_test, y_train, y_test = train_test_split(
        exames, diagnostico, test_size=0.3
    )

    classificador = RandomForestClassifier(n_estimators=100)
    classificador.fit(X_train, y_train)
    print(
        "Resultado da classificação %.2f%%"
        % (classificador.score(X_test, y_test) * 100)
    )
```

```python
matriz_corr_1 = matriz_corr[matriz_corr > 0.99]
matriz_corr_1
```

```python
matriz_corr_2 = matriz_corr_1.sum()
matriz_corr_2
```

```python
var_corr = matriz_corr_2[matriz_corr_2 > 1]
var_corr
```

```python
exames_4 = exames_3.drop(columns=var_corr.keys())
exames_4
```

```python
classificar(exames_4)
```

```python
exames_5 = exames_3.drop(columns=["exame_3", "exame_24"])
classificar(exames_5)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

k_best = SelectKBest(chi2, k=5)

exames_6 = exames_1.drop(columns=(["exame_4", "exame_29", "exame_3", "exame_24"]))

X_train, X_test, y_train, y_test = train_test_split(
    exames_6, diagnostico, test_size=0.3
)

k_best.fit(X_train, y_train)
k_best_train = k_best.transform(X_train)
k_best_test = k_best.transform(X_test)

k_best_test.shape
```

```python
classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(k_best_train, y_train)
print(
    "Resultado da classificação %.2f%%"
    % (classificador.score(k_best_test, y_test) * 100)
)
```

```python
from sklearn.metrics import confusion_matrix

conf_matrix = confusion_matrix(y_test, classificador.predict(k_best_test))

plt.figure(figsize=(7, 4))
sns.set(font_scale=2)
sns.heatmap(conf_matrix, annot=True, fmt="d").set(xlabel="Predição", ylabel="Real")
```

```python
from sklearn.feature_selection import RFE

X_train, X_test, y_train, y_test = train_test_split(
    exames_6, diagnostico, test_size=0.3
)

rfc_3 = RandomForestClassifier(n_estimators=100, random_state=SEED)
rfc_3.fit(X_train, y_train)

rfe = RFE(estimator=classificador, n_features_to_select=2, step=1)
rfe.fit(X_train, y_train)
rfe_train = rfe.transform(X_train)
rfe_test = rfe.transform(X_test)
classificador.fit(rfe_train, y_train)

conf_matrix = confusion_matrix(y_test, classificador.predict(rfe_test))
plt.figure(figsize=(7, 4))
sns.set(font_scale=2)
sns.heatmap(conf_matrix, annot=True, fmt="d").set(xlabel="Predição", ylabel="Real")

print(
    "Resultado da classificação %.2f%%" % (classificador.score(rfe_test, y_test) * 100)
)
```

```python
from sklearn.feature_selection import RFECV

X_train, X_test, y_train, y_test = train_test_split(
    exames_6, diagnostico, test_size=0.3
)

rfc_3 = RandomForestClassifier(n_estimators=100, random_state=SEED)
rfc_3.fit(X_train, y_train)

rfe_cv = RFECV(estimator=rfc_3, cv=5, step=1, scoring="accuracy")
rfe_cv.fit(X_train, y_train)
rfe_cv_train = rfe_cv.transform(X_train)
rfe_cv_test = rfe_cv.transform(X_test)
rfc_3.fit(rfe_cv_train, y_train)

conf_matrix = confusion_matrix(y_test, rfc_3.predict(rfe_cv_test))
plt.figure(figsize=(7, 4))
sns.set(font_scale=2)
sns.heatmap(conf_matrix, annot=True, fmt="d").set(xlabel="Predição", ylabel="Real")

print("Resultado da classificação %.2f%%" % (rfc_3.score(rfe_cv_test, y_test) * 100))
```
